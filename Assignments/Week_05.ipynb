## Vary the Ridge parameter `alpha` from 0 to any value you'd like above 1. Plot the Train and Test error. Describe what you see based on the `alpha` parameter's stiffness.

```python
#Ideal split 0.5, Small Alpha
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.5) 

model = Ridge(alpha = 0.0005)
model.fit(x_train, y_train)
model.coef_, model.intercept_

print('Test_0.5_A_Small =', mean_squared_error(y_test, np.dot(x_test, model.coef_) + model.intercept_))

print('Train_0.5_A_Small =', mean_squared_error(y_train, np.dot(x_train, model.coef_) + model.intercept_))

RTest_5_A_Small =  mean_squared_error(y_test, np.dot(x_test, model.coef_) + model.intercept_)
RTrain_5_A_Small = mean_squared_error(y_train, np.dot(x_train, model.coef_) + model.intercept_)

RTest_Size_5_A_Small = RTest_5_A_Small - RTrain_5_A_Small 
print('RTest_Size_5_A_Small', RTest_Size_5_A_Small)
```
